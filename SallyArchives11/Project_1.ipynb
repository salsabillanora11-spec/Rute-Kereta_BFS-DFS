```python
graph = {
      'A' : ['B', 'D'],
      'B' : ['A', 'E'],
      'C' : ['D', 'H'],
      'D' : ['A', 'C', 'I'],
      'E' : ['B', 'J', 'L'],
      'F' : ['G', 'N'],
      'G' : ['D', 'K'],
      'H' : ['E', 'L'],
      'I' : ['E', 'M'],
      'J' : ['F', 'N'],
      'K' : ['G', 'O'],
      'L' : ['H'],
      'M' : ['I'],
      'N' : ['J'],
      'O' : ['K']
}
```

```python
def bfs (graph, start):
      visited = {v: False for v in graph}

      queue = deque()
      route = []

      queue.append(start)
      visited[start]= True

      start_time = time.time()
      while queue:
            vertex = queue.popleft()
            route.append(vertex)

            for neighbor in graph[vertex]:
                  if not visited[neighbor]:
                        visited[neighbor] = True
                        queue.append(neighbor)
      
```
